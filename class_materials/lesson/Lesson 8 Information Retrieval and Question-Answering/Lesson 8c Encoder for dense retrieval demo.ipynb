```python
# documents
docs = [
    "Supervised learning trains algorithms using datasets where each input is paired with the correct output label.",
    "Machine learning models improve their performance by identifying patterns in large collections of data.",
    "Artificial intelligence has become widely used in applications such as recommendation systems and chatbots.",
    "The Amazon rainforest contains one of the most diverse ecosystems on Earth."
]

# compute embeddings for documents
d_emb = model.encode(docs)
```

```python
# query
query = "how to train a model using labeled examples"
q_emb = model.encode([query])
# cosine similarity
scores = cosine_similarity(q_emb, d_emb)[0]

# print results
for doc, score in sorted(zip(docs, scores), key=lambda x: -x[1]):
    print(f"{score:.3f}  {doc}")
```
